```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pickle
import h5py
import torch
import data_io
```

```python
dataset = h5py.File('/project/spice/radiation/ML/CRM/data/models/datain/train_data_021501AQ1H.hdf5', 'r')
```

```python
std_mx = data_io.NormalizersData('/project/spice/radiation/ML/CRM/data/models/normaliser/021501AQ1H_standardise_mx')
```

```python
q_phys = dataset['q_phys_train']
t_phys = dataset['t_phys_train']
q_tot = dataset['q_tot_train']
q_adv = dataset['q_adv_train']
theta = dataset['air_potential_temperature_train']
theta_adv = dataset['t_adv_train']
```

```python
q_phys_norm = std_mx.normalise(q_phys[:2000], std_mx.qphys_mean_np, std_mx.qphys_stdscale_np)
t_phys_norm = std_mx.normalise(t_phys[:2000], std_mx.tphys_mean_np, std_mx.tphys_stdscale_np)
q_tot_norm = std_mx.normalise(q_tot[:2000], std_mx.q_mean_np, std_mx.q_stdscale_np)
q_adv_norm = std_mx.normalise(q_adv[:2000], std_mx.qadv_mean_np, std_mx.qadv_stdscale_np)
theta_norm = std_mx.normalise(theta[:2000], std_mx.t_mean_np, std_mx.t_stdscale_np)
theta_adv_norm = std_mx.normalise(theta_adv[:2000], std_mx.tadv_mean_np, std_mx.tadv_stdscale_np)
```
